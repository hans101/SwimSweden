```python
import html5lib
from bs4 import BeautifulSoup
import pandas as pd
import requests
import re
from selenium import webdriver
from selenium.webdriver.support.ui import WebDriverWait, Select
from selenium.webdriver.support import expected_conditions as EC
from selenium.common.exceptions import TimeoutException
from selenium.webdriver.common.by import By
import time
```

```python
id_list = range(6372, 7000,1)
```

```python
browser = webdriver.Chrome(executable_path='D:/download/chromedriver.exe')
#browser.implicitly_wait(7)
```

```python
for id in id_list:
    url = 'http://www.octoopen.se/index.php?r=competition/view&id=%s&pageSize=2000&ResultComp_page=1' %id
    browser.get(url)
    timeout=10
    try:
        WebDriverWait(browser, timeout).until(EC.visibility_of_element_located((By.XPATH, "//div[@class='sixteen columns']")))
    except TimeoutException:
        print('problem')
        browser.quit()

    #old_value = browser.find_element_by_xpath("//*[@id='competitionview-grid']/table/tbody/tr[1]/td[1]")
    try: 
        select = Select(browser.find_element_by_id('ResultComp_class'))
        select.select_by_visible_text('Damer')
        browser.find_element_by_name('yt0').click()
        time.sleep(10)

        html = browser.page_source
        soup = BeautifulSoup(html, 'lxml')
        new_soup =str(soup.find_all('div', class_='sixteen columns'))

        df = pd.read_html(html, )[0]
        df.drop(df.columns[6], axis=1, inplace= True)
        df.Gren = df.Gren.astype('category')
        df.Namn = df.Namn.str.rstrip('^Swimmer$')

            # adding columns with information about pool_size, city, organizer 
        new_soup =str(soup.find_all('div', class_='sixteen columns'))
        df['comp_name'] = re.search('(?<=(<h2>))+.*(?=</)', new_soup).group(0)
        df['place'] = re.search('(?<=(Ort: ))+.*(?=(<b))', new_soup).group(0)
        df['date'] = re.search('(?<=(Datum: ))+.*(?=(<b))', new_soup).group(0)
        #df['date'] = pd.to_datetime(df['date'])
        df['organizer'] = re.search('(?<=(Arrangör: ))+.*(?=(<b))', new_soup).group(0)
            #pool size 50m or 25m
        if  (str(re.search('(?<=(Bassäng: ))+.*(?=(\n))', new_soup).group(0)) =='50 M'):
            df['pool_size'] = 1
        else:
            df['pool_size'] = 0

        df.to_csv('w%s.csv' %id)
    except:
        print('problem nr', id)
```
